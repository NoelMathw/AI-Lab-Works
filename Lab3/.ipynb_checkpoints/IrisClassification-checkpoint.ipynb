```python
import matplotlib . pyplot as plt
from sklearn . decomposition import PCA
from sklearn . datasets import load_iris
from sklearn.neural_network import MLPClassifier
from sklearn . model_selection import train_test_split
from sklearn.metrics import confusion_matrix
```

```python
# Load the Iris data .
data = load_iris ()
inputs = data.data
targets = data.target
```

```python
# Project the data into two dimensions using PCA .
pca = PCA(n_components=2)
compressed = pca.fit_transform(inputs)
```

```python
xtrain , xtest , ttrain , ttest = train_test_split(compressed,targets)

classifier = MLPClassifier()
classifier.fit(xtrain,ttrain)
```

```python
plt.figure()
plt.scatter(compressed[:,0],compressed[:,1],c=targets,cmap='coolwarm')
plt.colorbar()
plt.show()
```

```python
print("Training data score:",classifier.score(xtrain, ttrain),"Test data score:",classifier.score(xtest, ttest))
```

```python
predictdata = classifier.predict(xtest)

plt.figure()
plt.imshow(confusion_matrix(ttest, predictdata))
plt.colorbar()
plt.show()
```

<h1>Without using PCA</h1>

```python
myclassifier = MLPClassifier()
mxtrain,mxtest,mytrain,mytest = train_test_split(inputs,targets)
myclassifier.fit(mxtrain,mytrain)

mypredictdata = myclassifier.predict(mxtest)

plt.figure()
plt.imshow(confusion_matrix(mytest, mypredictdata))
plt.colorbar()
plt.show()
```
