```python
def Fibo(N):
    FiboList=[0]
    temp=1
    for i in range(0,N):
        FiboList.append(temp)
        temp = temp + FiboList[i]
    return FiboList;

print(Fibo(10))
```

```python
Fibo(18)
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

x = []
for i in range(0,10):
    x.append(i)

plt.figure(figsize=(5,7))

plt.scatter(x,Fibo(9),s=15,color="Red")
plt.title("Graph showing corresponding Fibonacci value (y-axis) at xth element")
plt.xlabel("Element number")
plt.ylabel("Fibonacci value")
plt.minorticks_on()
plt.tick_params(axis='both', which='major', length=5)
plt.axis([0, 10, 0, 35])
plt.grid(True)

# Set major ticks to differ by 1
plt.gca().xaxis.set_major_locator(ticker.MultipleLocator(1))
plt.gca().yaxis.set_major_locator(ticker.MultipleLocator(1))


plt.show()
```

```python
x_values = list(range(11))
squares = [x**2 for x in x_values]
cubes = [x**3 for x in x_values]
plt.plot(x_values, squares, c='blue')
plt.plot(x_values, cubes, c='red')
plt.axis([0, 11, 0, 1100])
plt.fill_between(x_values, cubes, squares, facecolor='blue', alpha=0.25)
plt.show()
```
