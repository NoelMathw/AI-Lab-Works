```python
import numpy as np
import matplotlib.pyplot as plt
print(np.random.rand,__doc__)
```

```python
uniform = np . random . rand ( 100 ) # Generate 100 uniform random points .
normal = np . random . randn ( 100 ) # Generate 100 normal random points .
```

```python
print(uniform)
```

```python
print(normal)
```

```python
plt.scatter(uniform, np.zeros(len(uniform)))
```

```python
plt.scatter(normal, np.zeros(len(normal)))
```

```python
plt.boxplot([uniform,normal])

plt.figure()
plt.boxplot([uniform ,normal])
plt.xticks([1 , 2], [" Uniform ", " Normal "])
plt.ylabel("$p(x)$")
plt.savefig("boxplot.png", bbox_inches ="tight")
plt.show()
```

```python
x = [1,2,3,4]
y = [10,20,25,30]
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(x,y,color="red",linewidth=1)
ax.scatter([2,3,4,5,6,7,8,9,21],[5,4,3,5,76,67,4,32,5],marker='.')
plt.show()
```
