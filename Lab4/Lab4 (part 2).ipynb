```python
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
```

```python
def train(x_train, y_train):
  inputs = tf.keras.Input(shape=(784,))

  dense = layers.Dense(64 ,activation ="relu")
  x = dense (inputs)
  x = layers.Dense(64, activation ="relu")(x)
  outputs = layers.Dense(10)(x)

  model = tf.keras.Model(inputs=inputs, outputs=outputs, name="fashion_model")
  model.summary()

  model.compile(
    loss = tf.keras.losses.SparseCategoricalCrossentropy(from_logits = True),
    optimizer = tf.keras.optimizers.RMSprop(),
    metrics =["accuracy"],)

  model.fit(x_train, y_train, batch_size = 64, epochs = 3, validation_split = 0.2)
  return model
```

```python
(x_train,y_train),(x_test, y_test) = tf.keras.datasets.fashion_mnist.load_data()

x_train = x_train.reshape(60000,784).astype("float32")/255
x_test = x_test.reshape (10000,784 ).astype ("float32")/ 255
```

```python
def plot_cf(model,y_test,x_test):
  predictions = model.predict(x_test).argmax(axis=1)
  plt.figure()
  plt.imshow(confusion_matrix(y_test, predictions))
  plt.show()
```

```python
mymodel = train(x_train,y_train)

plot_cf(mymodel,y_test,x_test)
```
